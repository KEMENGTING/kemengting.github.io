## 10- Top 5 highest OBP (on base percentage) with at least 500 PAs in 1977?  (Dataframe, first name, last name, OBP, ordered descending by OBP)

```python
df10 = pd.read_csv('Batting.csv',usecols=['nameFirst', 'nameLast','yearID','H','AB','BB','HBP','SH','SF'])
A10 = df10[df10['yearID'] == 1977].groupby(['nameFirst','nameLast','yearID']).sum()
A10['PA'] = A10['AB'] + A10['BB'] + A10['HBP'] + A10['SH'] + A10['SF']
A10['OBP'] = (A10['H'] + A10['BB'] + A10['HBP']) / (A10['AB'] + A10['BB'] + A10['HBP'] + A10['SF'])
A10[A10['PA']>= 500].sort_values(by ='OBP',ascending=False).head()
```

## 13-	Who led MLB with the most hits the most times?  And how many times?  (Dataframe, Number of hits)

## 24- Who is the youngest player to hit 50 or more HRs in a single season? (Dataframe, first name, last name, HRs, season)

Prince Fielder

```python
df24 = pd.read_csv('Batting.csv',usecols=['nameFirst','nameLast','birthYear','yearID','HR'])
A24 = df24.groupby(['nameFirst','nameLast','birthYear','yearID'], as_index = False).sum()
AA24 = A24[A24['HR']>= 50]
AA24['differ'] = AA24['yearID'] - AA24['birthYear']
min(AA24['differ'])
AA24[AA24['differ']== 23][['nameFirst','nameLast','HR','yearID']]
```

## 25- Who are the five youngest players to hit 300 or more HRs for their career? (Dataframe, first name, last name, season they eclipsed more than 300 HRs)

```python
df25 = pd.read_csv('Batting.csv',usecols=['playerID','nameFirst','nameLast','birthYear','yearID','HR'])
A25 = df25.groupby(['playerID','nameFirst','nameLast','birthYear','yearID'], as_index = False).sum()
A25['cumsumHR'] = A25.groupby('playerID', as_index=False)['HR'].cumsum()
AA25 = A25[A25['cumsumHR']>= 300].groupby('playerID').first()
AA25['differ'] = AA25['yearID'] - AA25['birthYear']
min(AA25['differ'])
AA25[AA25['differ']== 28][['nameFirst','nameLast','yearID']]
```
